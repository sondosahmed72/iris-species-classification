```python

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))
for i, col in enumerate(numeric_features):
    sns.histplot(df[col], kde=True, ax=axes[i])
    axes[i].set_title(col)
plt.show()
```

```python
def quantile_transform(data,cols):
    qt = QuantileTransformer(output_distribution='normal')
    for col in cols:
        data[col] = qt.fit_transform(data[[col]])
    return data

df = quantile_transform(df,numeric_features) 
```

```python

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))
for i, col in enumerate(numeric_features):
    sns.histplot(df[col], kde=True, ax=axes[i])
    axes[i].set_title(col)
plt.show()
```

* transformation

```python
from sklearn.preprocessing import LabelEncoder

# Encoding the 'Class' column
label_encoder = LabelEncoder()
df['Class'] = label_encoder.fit_transform(df['Class'])
df['Class'].unique()  # Checking the encoded classes
```

# Build Model

```python
X=df.drop(columns=['Class'])
y=df['Class']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
models = {
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=10000, random_state=42),
    'Support Vector Machine': SVC(),
}
```

```python
def evaluate_model(model,X_trian,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    accuracy=accuracy_score(y_test,y_pred)
    precision = precision_score(y_test, y_pred,average='weighted' )
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    M = confusion_matrix(y_test, y_pred)
    return accuracy, precision, recall, f1 ,M
```

```python
performance = {}
for name, model in models.items():
        accuracy, precision, recall, f1 , m = evaluate_model(model, X_train, X_test, y_train, y_test)
        performance[name] = {'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1 Score': f1}
        plt.figure(figsize=(8, 6))
        sns.heatmap(m, annot=True, cmap='Blues', fmt='g', cbar=False)
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.title('Confusion Matrix - K-Nearest Neighbors')
        plt.show()
```

```python
performance_df = pd.DataFrame(performance).T
performance_df
```

# Prediction

```python
# Example: Predict the class for a new sample
new_sample = [[5.1, 3.5, 1.4, 0.2]]  # Example new iris flower features
new_sample_scaled = scaler.transform(new_sample)

# Choose the best model (assuming Random Forest in this case)
best_model = models['Support Vector Machine']
predicted_class = best_model.predict(new_sample_scaled)
print(f'The predicted class for the new sample is: {predicted_class[0]}')
```
